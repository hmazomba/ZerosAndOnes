**Retrieving a dataset from Quandl**

```
!pip install quandl
import quandl
QUANDL_API_KEY = 'YesoVjvNS1D7c5jWycBa'
quandl.ApiConfig.api_key = QUANDL_API_KEY
df = quandl.get('EURONEXT/ABN')
df.head()
```

```
%matplotlib inline
import matplotlib.pyplot as plt
df.plot()
```

**Plotting price and volume**

```
prices = df['Last']
volumes = df['Volume']

#Plot consisting of daily closing prices
priceplot = plt.subplot2grid((4, 4), (0, 0), rowspan=3, colspan=4)
priceplot.plot(prices.index, prices, label='Last')
plt.title('ABN Last Price from 2016-2020')
plt.legend(loc=2)

#Plot consisting of daily trading volume
volumeplot = plt.subplot2grid((4, 4), (3,0), rowspan=1, colspan=4)
volumeplot.bar(volumes.index, volumes)
plt.title('ABN Daily Trading Volume')
plt.gcf().set_size_inches(9, 6)
plt.subplots_adjust(hspace=0.75)
```
